```python
import numpy as np
import pandas as pd
import csv
from tqdm import tqdm
import math
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

with open("winequality-red.csv", 'r') as f:
    wines = list(csv.reader(f, delimiter=";"))

wines_header = np.array(wines[0])  # with label header
wines = np.array(wines[1:], dtype=np.float) # with label
```

```python
# breast-cancer-wisconsin
with open("breast-cancer-wisconsin.data", 'r') as f:
    tumors = list(csv.reader(f, delimiter=";"))
    
tumors_header = ["id","clump thickness","cell size","cell shape","marginal adhesion",\
                 "single epithelial cell size","number of bare nuclei","bland chromatin",\
                 "number of normal nuclei","mitosis","class"] # with label header
#print(len(tumors))
#tumors = np.array(tumors[0:]) # with label
invalid_index = []
#print(tumors)
for i in tqdm(range(len(tumors))):
    tumors[i] = tumors[i][0].split(",")
    for j in range(len(tumors[i])):
        if tumors[i][j].isnumeric() == False:
            invalid_index.append(i)    #the whole row
    if int(tumors[i][-1]) <= 2:
        tumors[i][-1] = '0'
    else:
        tumors[i][-1] = '1'
       
invalid_index.sort(reverse=True)

for i in invalid_index:
    tumors.remove(tumors[i])

tumors = np.array(tumors[0:],dtype=np.int)

# Here is the place to return the numpy array to model trainers
print(tumors_header)
```

```python
# print(len(tumors))
# tumors_header.remove('cell shape')
# tumors = np.delete(tumors,3,axis=1)

df = pd.DataFrame(tumors[:,1:],columns=tumors_header[1:])
mask = np.zeros_like(df.corr(), dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

plt.subplots(figsize = (15,15))
sns.heatmap(df.corr(), 
            annot=True,
            mask = mask,
            cmap = 'RdBu_r',
            linewidths=0.1, 
            linecolor='white',
            vmax = .9,
            vmin = .1,
            square=True)
plt.title("Tumor Dataset Feature Correlation Analysis", y = 1.03,fontsize = 15)
plt.savefig("tumors_corr.png")
```

```python
pos_index = []
neg_index = []
for index in range(len(tumors)):
    if tumors[index][-1] == 1:
        pos_index.append(index)
    else:
        neg_index.append(index)

pos_tumors = tumors[pos_index]
neg_tumors = tumors[neg_index]
print(len(neg_tumors))
print(len(pos_tumors))


labels = 'Malignant (444)', 'Benign (239)'
sizes = [444,239]
colors = ['gold','lightcoral']
explode = (0.1, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')
plt.savefig("tumors_number.png")
```
